# Generate masks

## Use a circle to represent a spot

With the tool **labelme**, I was able to annotate spot on the salmon image with a circle which is describled by two pixels. One pixel is the circle center (center pixel). The other pixel is on the circle edge (edge pixel). To create masks, we need to find out all the pixels that are located within the annotated circels.

The function below extracts and parses all the circels from the json file and return a data frame. The first two columns are the coordinates of the center pixels. The last column are radius.

```python
def extract_circles_from_labelme_json(json_path):
    '''
    Extract center and radious of circles from a json mark file generated with labelme 
    '''
    
    with open(json_path) as f:
        data = json.load(f)
    # P1: circle center; P2: point on circle
    P1 = [p['points'][0] for p in data['shapes']]
    P2 = [p['points'][1] for p in data['shapes']]
    
    # calculate radius
    x_delta = (np.array(P2) - np.array(P1))[:, 1]
    y_delta = (np.array(P2) - np.array(P1))[:, 0]
    R = list(map(math.hypot, x_delta, y_delta))
    
    circles = pd.DataFrame(P1, columns=['y', 'x'])
    circles['r'] = R
    
    return circles
```

```python
circles = extract_circles_from_labelme_json('train_annotation/1708cs.json')
circles.head()
```

## Find pixels within circles

**How to find all pixels within a circle?**
We calculate the distance between the center pixel and the edge pixel. This distance is the radius of the circle. If the distance between a pixel and the center pixel is less than the radius, then this pixel is within the circle.

The image size is 1300 x 900 = 1170000. If we calculate the distance of each of the 1,170,000 pixels, it will be very computationally intensive. Instead, we should identify pixels around the circle and only examine those pixels. Such pixels can be identify by drawing a square outside the circle. The square center is the same as the circle center. The square size is **2\*radius by 2\*radius**.

```python
def identify_pixels_within_circle(y, x, r):
    '''
    Identify pixels within circle given the circle center coordinate and radius.
    
    :param x: x coordinate of a circle
    :param y: y coordinate of a circle
    :param r: the radius of a circle
    
    :return: pixel coordinates of the square
    '''
    x_arr = list(range(int(x - r), int(x + r)))
    y_arr = list(range(int(y -r ), int(y + r)))
    xx, yy = np.meshgrid(x_arr, y_arr)
    xx = xx.flatten()
    yy = yy.flatten()
    pixels = list(np.transpose(np.vstack((xx, yy))))
    
    new_pixels = []
    for i in range(len(pixels)):
        px, py = pixels[i]
        if math.hypot(px-x, py-y) < r:
            new_pixels.append([px, py])
    return np.array(new_pixels)
```

```python
def generate_mask(image_path, json_path):
    '''
    Generate mask given the image and the annotation json file for that image.
    '''
    
    img = io.imread(image_path)
    circles = extract_circles_from_labelme_json(json_path)
    
    # inital pixels with zeros
    mask = np.zeros(img.shape)
    # loop through each circles
    for i in range(len(circles)):
        x, y, r = circles.loc[i]
        # if the pixel is within the circle, change the value to 1.
        pixels = identify_pixels_within_circle(x, y, r) 
        mask[pixels[:, 0], pixels[:, 1]] = 1
        
    return mask
```

## Image and mask example

```python
fig, (ax0, ax1, ax2, ax3) = plt.subplots(1, 4, figsize=(18, 12))
ax0.imshow(plt.imread('train/1707cs.jpg'))
ax1.imshow(generate_mask('train/1707cs.jpg', 'train_annotation/1707cs.json'))
ax2.imshow(plt.imread('train/1708cs.jpg'))
ax3.imshow(generate_mask('train/1708cs.jpg', 'train_annotation/1708cs.json'))
plt.show()
```
